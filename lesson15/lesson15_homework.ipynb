```python
import matplotlib.pyplot as plt

# 資料
labels = ['Nokia', 'Samsung', 'Apple', 'Lumia']
values = [20, 30, 45, 10]
colors = ['yellow', 'green', 'red', 'blue']
explode = (0.3, 0, 0, 0)  # only "explode" the first slice (Nokia)

# 繪製 Pie 圖表
plt.pie(values,
        labels=labels,
        colors=colors,
        autopct='%1.1f%%',
        startangle=185,
        explode=explode)  # add explode here

# 設定圖表屬性
plt.axis('equal')  # 讓圖表呈現為正圓形

# 顯示圖表
plt.show()
```
