```python
import matplotlib.pyplot as plt


labels = ['Nokia', 'Samsung', 'Apple', 'Lumia']
values = [20, 30, 45, 10]
colors = ['yellow', 'green', 'red', 'blue']
explode = (0.3, 0, 0, 0)  

figure = plt.figure(figsize=(5,5),dpi=150,)
axes = figure.add_subplot(1,1,1)
axes.pie(values,
        labels=labels,
        colors=colors,
        autopct='%1.1f%%',
        startangle=180,
        explode=explode,
        shadow=True)

plt.show()
```
